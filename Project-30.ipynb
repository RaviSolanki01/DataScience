### What is the major reason people being kidnapped in each and every state?

```python
from  sklearn.impute import SimpleImputer
```

```python
imp_zero = SimpleImputer(missing_values=np.nan,fill_value=0.0)
```

```python
for col in data.columns:
    if data[col].dtype == 'float64':
        data[col] = imp_zero.fit_transform(data[col].values.reshape(-1,1))
```

```python
for col in data.columns:
    if data[col].dtype == 'float64':
        data[col]=(data[col].astype('int64'))
```

```python
data.info()
```

```python
cat_data = [x for x in data.dtypes.index if data.dtypes[x]==object]
```

```python
for col in cat_data:
    print('\nFrequency of Categories for varible %s'%col)
    print(data[col].value_counts())
```

```python
plt.figure(figsize=(25,10))
sns.catplot(y="Area_Name", x="K_A_Grand_Total", kind="bar", data=data)
```

```python
data.groupby(by=['Area_Name','Group_Name'], as_index=False)['K_A_Grand_Total'].count().sort_values(by='K_A_Grand_Total', ascending=False).head()
```

```python
data.groupby(by=['Area_Name','Group_Name'], as_index=False)['K_A_Female_Total'].count().sort_values(by='K_A_Female_Total', ascending=False).head()
```

```python
data.groupby(by=['Area_Name','Group_Name'], as_index=False)['K_A_Male_Total'].count().sort_values(by='K_A_Male_Total', ascending=False).head()
```
